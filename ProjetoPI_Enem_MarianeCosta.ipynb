```python
import pandas as pd
fonte = 'https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true'
dados_enem = pd.read_csv(fonte) 
dados_enem.head(100) #mostramos os dados dos 100 primeiros participantes
```

#Média de idades dos participantes

*Armazenamos as idades em um vetor e em seguida calculamos a média.*

#Criando funções para o cálculo da média

**Optamos restringir a quantidade de idades avaliadas para 20 para facilitar uma posterior verificação dos cálculos.**



```python
#Carregar as Idades a partir do arquivo CSV
def CarregaDadosEnem(Vet,dados_enem):
  for ind in dados_enem.index:  
    dadosEnem = dados_enem.loc[ind]  
    Vet.append(dadosEnem.NU_IDADE)
```

```python
#Exibir as idades 
def mostraIdade(Vet):
  for ind in range(20):
    print(Vet[ind], end='|')
```

```python
#Cálculo da Média das 20 primeiras idades
def calcMedia(Vet, dados_enem):
    mediaIdade = 0
    for ind in dados_enem.index:  
      dadosEnemIdade = dados_enem.loc[ind]  
      Vet.append(dadosEnemIdade.NU_IDADE)
    soma = 0
    for ind in range (20):
      soma = soma + Vet[ind]
    mediaIdade = soma/20  
    print(f"{mediaIdade:.2f}")   
```

#Exibindo as 20 primeiras idades da Base de Dados

```python
Vet = []
CarregaDadosEnem(Vet, dados_enem)
mostraIdade(Vet)
```

# Resultado da Média das 20 primeiras idades

```python
Vet = []
calcMedia(Vet, dados_enem)
```

```python
def encontrarAlunoComMaiorNotaNaRedacao(dados):
  maior_nota = 0
  for i in dados.index:
    nota = dados.loc[i, 'NU_NOTA_REDACAO']
    if maior_nota < nota:
      maior_nota = nota

  return dados.loc[dados['NU_NOTA_REDACAO'] == maior_nota]
```

```python
def OpcoesDisponiveis():
  print('*'*50)
  print('Bem vindo! \nPor favor selecione uma das opções a seguir:')
  print('\nDigite F para selecionar o gênero Feminino.')
  print('Digite M para selecionar o gênero Masculino.')
  print('Digite [S/N] para dar continuidade ou encerrar o programa. \n')
  print('*'*50)
```

```python
def generoParticipantes(municipio_filtrado, listaGeneros):
  for ind in municipio_filtrado.index:  
        dadosGenero = municipio_filtrado.loc[ind]  
        listaGeneros.append(dadosGenero.TP_SEXO)
```

```python
#----Programa Principal----
OpcoesDisponiveis()
listaGeneros = []
contFem = 0
contMasc = 0
generoParticipantes(municipio_filtrado, listaGeneros)
for genero in listaGeneros:
  if genero in 'Ff':
    contFem += 1
  else:
    contMasc += 1

while True:
  pergunta = str(input('Digite um gênero [F/M]: \n'))
  if pergunta == 'F' or pergunta == 'f':
    print('Você digitou F. O número de participantes que se identificam com o gênero feminino é:', contFem)
  elif pergunta == 'M' or pergunta == 'm':
    print('Você digitou M. O número de participantes que se identificam com o gênero masculino é:', contMasc)
  else:
    print('[ERRO] Por favor, redigite. \n')
    #OpcoesDisponiveis()
  fim = str(input('Deseja continuar? [S/N] \n'))
  if fim in 'Nn':
    print('\n Tchau! :)')
    break
  if fim in 'Ss':
    continue 
```
